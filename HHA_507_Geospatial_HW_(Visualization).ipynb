# Geospatial Visualization

```python
df_gpd = gpd.read_file('/content/Cancer_Rates.geojson')
df_gpd
```

This dataset is of different Cancer Rates in Lake County, Illinois. It looks at differnce cancers such as Colorectal,	Lung_Bronc,	Breast_Can,	Prostate_C,	Urinary_Sy, and	All_Cancer.

```python
df_gpd['All_Cancer'] = (df_gpd['All_Cancer']/ 1000)
```

The column of interest would be 'All_Cancer' since it is the cullmination of all the cancer rates. Since the rate is per 100,000 persons I am dividing it by 1000 in order to give me the percentage to make it easier to interpret.

```python
df_gpd.plot('All_Cancer', legend=True)
```

This is the visualization of the data using the longitue and latitude coordinates in the 'geometry' column. As can be seen the cancer rates are higher in a particular zip code to the south of the map colored in yellow at around 4.5%.

```python
df_gpd3 = gpd.read_file('/content/HealthInsuranceCoverage.geojson')
df_gpd3
```

This next dataset is based from the Detroit Tri-County region and collected information on what parts of the popluation had insurance coverage.

```python
df_gpd5 = gpd.read_file('/content/Mortality_Rates.geojson')
df_gpd5
```

The dataset called Mortality Rates measures the mortality rates for Lake County, Illinois. It measures differnect statistics such as PCT_EARLY, Avg_Age,	Cancer_D,	Heart_D,	and COPD_D.
